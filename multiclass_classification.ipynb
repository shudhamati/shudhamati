```python
df.drop(df[(df['total_fiat'])==-1].index,inplace=True)
df.drop(df[(df['total_biat'])==-1].index,inplace=True)
df['min_active']=df.replace(-1,df['min_active'].median())
df['max_active']=df.replace(-1,df['max_active'].median())
df['min_idle']=df.replace(-1,df['min_idle'].median())
df['max_idle']=df.replace(-1,df['max_idle'].median())
```

```python
dummies=pd.get_dummies(df['class1'],drop_first=True)
```
